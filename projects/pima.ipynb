```python
import pandas as pd
import numpy as np


data = pd.read_csv('../data/pima.data',header=None)
data.head(5)
```

```python
# 利用map函数，将None替换为空值
data[1] = data[1].map(lambda x:np.nan if str(x)=='None' else x)
data.head(5)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# 删除有空值的行
data.dropna(subset=[1],inplace=True)
data.head(5)
```

```python
data.shape
```

```python
# 查看重复值情况
data.duplicated().sum()
```

```python
# 转换数据类型
data[1] = data[1].astype(np.int64)
data.head(5)
```

```python
# 筛选数据
data = data[(data[3]<80)&(data[2]>0)&(data[5]>0)]
data.head(5)
```
