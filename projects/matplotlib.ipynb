```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family']='simhei'
plt.rcParams['axes.unicode_minus']=False

subject=['语文','数学','英语','化学','地理']
score1=[88,94,92,83,78]
score2=[77,84,88,69,90]
```

```python
plt.plot(subject,score1,'o-',label='s1')
plt.plot(subject,score2,'d--',label='s2')
plt.legend()
plt.grid(axis='y',ls=':')
plt.ylim(60,100)
plt.show()
```

```python
plt.bar(np.arange(len(subject)),score1,width=0.4)
plt.bar(np.arange(len(subject))+0.4,score2,width=0.4)
plt.xticks(np.arange(len(subject))+0.4/2,subject,rotation=30)
plt.ylim(0,100)
plt.show()
```

```python
plt.bar(subject,score1)
plt.bar(subject,score2,bottom=score1)
plt.show()
```

```python
plt.barh(np.arange(len(subject)),score1,height=0.4)
plt.barh(np.arange(len(subject))+0.4,score2,height=0.4)
plt.yticks(np.arange(len(subject))+0.4/2,subject)
plt.show()
```

```python
plt.barh(subject,score1)
plt.barh(subject,score2,left=score1)
plt.show()
```

```python
err_value=[4,6,5,7,5]
err_style=dict(elinewidth=2,ecolor='black',capsize=3)
plt.bar(subject,score1,yerr=err_value,error_kw=err_style)
plt.show()
```

```python
explode=[0.01]*len(subject)
plt.pie(score1,explode,autopct='%.2f%%')
plt.legend(subject,bbox_to_anchor=[-0.7,0,0.5,1])
plt.show()
```

```python
explode=[0]*len(subject)
plt.pie(score1,explode,wedgeprops=dict(width=0.4,edgecolor='w'))
plt.legend(subject,bbox_to_anchor=[-1,0,0.5,1])
plt.show()
```

```python
color=['pink','yellow','c','green','orange']
explode=[0]*len(subject)
plt.pie(score1,explode,colors=color,wedgeprops=dict(width=0.7,edgecolor='w'))
plt.pie(score2,explode,colors=color,wedgeprops=dict(width=0.4,edgecolor='w'))
plt.legend(subject,bbox_to_anchor=[-1,0,0.5,1])
plt.show()
```

```python
title=['语文','数学','英语','化学','地理']
s1=[88,94,92,83,78]
s2=[77,84,88,69,90]

theta=np.linspace(0,2*np.pi,len(subject),endpoint=False)
theta=np.concatenate((theta,[theta[0]]))
title=np.concatenate((title,[title[0]]))
s1=np.concatenate((s1,[s1[0]]))
s2=np.concatenate((s2,[s2[0]]))


plt.polar(theta,s1)
plt.fill(theta,s1,alpha=0.4,label='s1')
plt.polar(theta,s2)
plt.fill(theta,s2,alpha=0.5,label='s2')

plt.thetagrids(theta*180/np.pi,title)
plt.legend(bbox_to_anchor=[-0.6,0,0.5,1])
plt.title('雷达图',fontsize=16)

plt.show()
```
