# 读取csv

```python
import pandas as pd
df = pd.read_csv(open('../data/renkou.csv'))
df.head(3)
```

```python
# read_table也可以读取，但是要指定分隔符
df = pd.read_table(open('../data/renkou.csv'),sep=',')
df.head(3)
```

```python
# 指定列作为索引
df = pd.read_csv(open('../data/renkou.csv'),index_col='total')
df.head(3)
```

```python
# 1.若csv没有表头，则默认分配
# df = pd.read_csv(open('../data/renkou.csv'),header=None)
# df.head(3)

# 2.也可以自己指定表头
df = pd.read_csv(open('../data/renkou.csv'),names=['年份','男','女','城镇','农村'])
df.head(3)
```

```python
# 通过skiprows函数跳过一些行
df = pd.read_csv(open('../data/renkou.csv'),skiprows=[0,6])
df.head()
```

```python
# 通过nrows函数只选取指定列
df = pd.read_csv(open('../data/renkou.csv'),nrows=5)
df
```

```python
# 只选取指定列
df = pd.read_csv(open('../data/renkou.csv'),usecols=['year','total'])
df.head(3)
```

```python
# 只选取指定行列
df = pd.read_csv(open('../data/renkou.csv'),nrows=5,usecols=['year','total'])
df
```
