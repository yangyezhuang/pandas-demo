```python
import pandas as pd
import numpy as np

data = {
    'name':['张三','李四','小明','王五'],
    'sex':['male','female','','male'],
    'year':[24,37,24,33],
    'city':['天津','深圳','','北京']
}

df = pd.DataFrame(data)
df
```

```python
# df.replace('','不确定')
df.replace('',np.nan)
```

```python
df.replace(['',24],['null',33])     # 以列表形式传参并替换
```

```python
df.replace({'':'null',24:33})     # 以字典形式传参并替换
```
