# 利用loc、iloc对行，列进行选取

|    |   索引 | 描述  |
| ---- |  ----  | ----  |
|  1  |  .loc()  | 按行索引标签选取数据 |
|  2  | .iloc()  | 按行索引位置选取数据基于标签 |

# .loc[]

Pandas提供了各种方法来完成基于标签的索引。 切片时，也包括起始边界。整数是有效的标签，但它们是指标签而不是位置。
.loc()具有多种访问方式，如 -

单个标量标签标签列表切片对象一个布尔数组
loc需要两个单/列表/范围运算符，用","分隔。第一个表示行，第二个表示列。

### loc函数：通过行索引 "Index" 中的具体值来取行数据（如取"Index"为"A"的行）

### iloc函数：通过行号来取行数据（如取第二行的数据）

使用行和列名称时选择特定行和/或列loc

使用表中的位置时选择特定行和/或列iloc

```python
# 只选取一行
df.loc[5]
```

```python
# 选取多行（选取第五行）

# df[:5]
df.loc[:5,:]
```

```python
df1 = df.set_index('area')
df1.loc[['滨湖区','梁溪区']]
```

## 2. 利用loc提取列数据

```python
# 选取area列
# ','前的部分表示取所有行，','后的部分表示取所有列，即取所有行的‘area’列

df.loc[:,['area']]
```

```python
# 选取多列
df.loc[:,['area','month_sale']]
```

```python
df.loc[[1,3,5,7,9],['month_sale']]
```

```python
# 选取 5行与 3列
df.loc[5,['area','month_sale']]
```

```python
# 选取指定行与列
df.loc[5:10,:]
```

```python
# 选取所有行与列
df.loc[:,:]
```

# .iloc[]

```python
# 选取前三行
df.iloc[:3]
```

```python
df.iloc[1:4]
```

```python
# 筛选第 3行到 6行和 0到 5列

df.iloc[3:6,0:3]
```

```python
df.iloc[[1,3,5,7,9],[0,1,2]]
```
